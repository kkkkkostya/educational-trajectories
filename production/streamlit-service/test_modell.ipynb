```python
import numpy as np
from carte_ai import CARTERegressor
from numpy.exceptions import AxisError

orig_predict = CARTERegressor.predict

def patched_predict(self, X):
    try:
        return orig_predict(self, X)
    except AxisError:
        print('WE HERE')
        # восстанавливаем скалярное предсказание
        return np.array([self.model_list_[0].predict(X)[0]])

CARTERegressor.predict = patched_predict

print(orig_predict)
# Теперь просто:
y_pred = estimator.predict(X_test)
```
